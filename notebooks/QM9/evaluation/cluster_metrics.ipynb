```python

import json
import pandas as pd
import plotly.graph_objects as go

from pathlib import Path
from IPython.display import display, Markdown
from plotly.subplots import make_subplots
```

```python
report_path = Path('results/cluster_reports/consolidated_report.json')
with report_path.open() as f:
    report_data = json.load(f)

cluster_score_keys = sorted(
    k for k in report_data.keys() if k.endswith('cluster_scores.json')
)

isomer_non_euclidean_keys = [
    k for k in cluster_score_keys if 'isomers_non_euclidean' in k
]
isomer_euclidean_keys = [
    k for k in cluster_score_keys if 'isomers_euclidean' in k
]
other_keys = [
    k for k in cluster_score_keys
    if k not in isomer_non_euclidean_keys and k not in isomer_euclidean_keys
]

len(cluster_score_keys)
```

```python
def _title_from_key(key: str) -> str:
    # e.g. non_euclidean/wasserstein_cluster_scores.json -> non_euclidean/wasserstein
    return key.replace('_cluster_scores.json', '')


def _display_scores(keys, section_title=None):
    if section_title:
        display(Markdown(f"# {section_title}"))

    if not keys:
        display(Markdown("No entries found."))
        return

    for key in keys:
        payload = report_data.get(key, {})
        title = _title_from_key(key)

        display(Markdown(f"## {title}"))

        summary = {
            "embedding": payload.get("embedding"),
            "best_k_used": payload.get("best_k_used"),
            "valid_count": payload.get("valid_count"),
            "total_count": payload.get("total_count"),
        }
        display(pd.DataFrame([summary]))

        congruence = payload.get("congruence")
        if isinstance(congruence, dict) and len(congruence) > 0:
            df_congruence = (
                pd.DataFrame(
                    [{"cluster": str(k), "congruence": v} for k, v in congruence.items()]
                )
                .sort_values("cluster")
                .reset_index(drop=True)
            )
            display(Markdown("**Congruence**"))
            display(df_congruence)
        else:
            display(Markdown("**Congruence**: not available"))

        mcs = payload.get("molecular_cluster_score")
        if isinstance(mcs, dict) and len(mcs) > 0:
            if "error" in mcs:
                display(Markdown(f"**Molecular Cluster Score**: {mcs['error']}"))
            else:
                df_mcs = pd.DataFrame([mcs])
                display(Markdown("**Molecular Cluster Score**"))
                display(df_mcs)
        else:
            display(Markdown("**Molecular Cluster Score**: not available"))


_display_scores(isomer_non_euclidean_keys, section_title="Isomer Non-Euclidean")
_display_scores(isomer_euclidean_keys, section_title="Isomer Euclidean")
_display_scores(other_keys, section_title="All Other Cluster Scores")
```

```python
def plot_all_categories(report_data, categories=None):
    rows = []
    metrics = ['structure_score', 'property_score', 'category_score', 'separation_score', 'structure_class_score']
    
    for key, payload in report_data.items():
        if '/' not in key:
            continue
            
        parts = key.split('/')
        if len(parts) < 2:
            continue
            
        category = parts[0]
        method = parts[1].replace('_cluster_scores.json', '')

        # Filter categories if provided
        if categories is not None and category not in categories:
            continue

        mcs = payload.get("molecular_cluster_score", {})
        
        if isinstance(mcs, dict) and all(m in mcs for m in metrics):
            row = {"category": category, "method": method}
            for m in metrics:
                row[m] = mcs.get(m, 0)
            rows.append(row)

    if not rows:
        print("No valid cluster scores found for the selected categories.")
        return

    df = pd.DataFrame(rows)

    # If categories not provided, infer from dataframe
    if categories is None:
        categories = df['category'].unique()
    else:
        categories = [c for c in categories if c in df['category'].unique()]

    fig = make_subplots(
        rows=len(categories), cols=1,
        specs=[[{'type': 'polar'}] for _ in categories],
        subplot_titles=[f"Category: {c}" for c in categories],
        vertical_spacing=0.05
    )

    colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A',
              '#19D3AF', '#FF6692', '#B6E880']

    for i, cat in enumerate(categories):
        cat_df = df[df['category'] == cat]

        for j, (_, row) in enumerate(cat_df.iterrows()):
            values = [row[m] for m in metrics]
            values.append(values[0])

            fig.add_trace(
                go.Scatterpolar(
                    r=values,
                    theta=metrics + [metrics[0]],
                    name=row['method'],
                    legendgroup=cat,
                    line=dict(color=colors[j % len(colors)])
                ),
                row=i+1, col=1
            )

    fig.update_layout(
        height=500 * len(categories),
        template="plotly_dark",
        showlegend=True,
        margin=dict(t=100, b=50, l=100, r=100),
        legend=dict(groupclick="toggleitem")
    )
    
    fig.update_polars(radialaxis=dict(visible=True, range=[0, 1]))
    fig.show()
```

```python
plot_all_categories(report_data, categories=['isomers', 'isomers_euclidean'])
```

```python
plot_all_categories(report_data, categories=['descriptors_kmeans', 'euclidean_metrics'])
```

```python
plot_all_categories(report_data, categories = ['non_euclidean', 'isomers_non_euclidean'])
```
